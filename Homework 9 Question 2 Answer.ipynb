```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.cluster import AgglomerativeClustering
from matplotlib import pyplot as plt
```

(a) Run k-Means algorithm using several values of k.

```python
data = pd.read_csv('hwk09.csv')
h1 = data['x'].values
h2 = data['y'].values
X = np.array(list(zip(h1, h2)))

fig = plt.figure(figsize=(10,10))

plt.subplot(221)
k = 1
kmeans = KMeans(n_clusters=k)
kmeans = kmeans.fit(X)
labels = kmeans.predict(X)

for i in range(k):
    pts = np.array([X[j] for j in range(len(X)) if labels[j] == i])
    plt.scatter(pts[:,0], pts[:,1], s=10, cmap='rainbow')
    plt.title('Number of clusters = {}'.format(k))
    
plt.subplot(222)
k = 2
kmeans = KMeans(n_clusters=k)
kmeans = kmeans.fit(X)
labels = kmeans.predict(X)

for i in range(k):
    pts = np.array([X[j] for j in range(len(X)) if labels[j] == i])
    plt.scatter(pts[:,0], pts[:,1], s=10, cmap='rainbow')
    plt.title('Number of clusters = {}'.format(k))
    
plt.subplot(223)
k = 3
kmeans = KMeans(n_clusters=k)
kmeans = kmeans.fit(X)
labels = kmeans.predict(X)

for i in range(k):
    pts = np.array([X[j] for j in range(len(X)) if labels[j] == i])
    plt.scatter(pts[:,0], pts[:,1], s=10, cmap='rainbow')
    plt.title('Number of clusters = {}'.format(k))
    
plt.subplot(224)
k = 4
kmeans = KMeans(n_clusters=k)
kmeans = kmeans.fit(X)
labels = kmeans.predict(X)

for i in range(k):
    pts = np.array([X[j] for j in range(len(X)) if labels[j] == i])
    plt.scatter(pts[:,0], pts[:,1], s=10, cmap='rainbow')
    plt.title('Number of clusters = {}'.format(k))
```

(b) Run DBSCAN algorithm with several pairs of  and minPts

```python
data = pd.read_csv('hwk09.csv')
h1 = data['x'].values
h2 = data['y'].values
X = np.array(list(zip(h1, h2)))

#NOTE: Outlier points displayed in black
fig = plt.figure(figsize=(10,10))

plt.subplot(221)
db = DBSCAN(eps=1, min_samples=3).fit(X)
labels = db.labels_
cr = np.zeros_like(labels, dtype = bool)
cr[db.core_sample_indices_] = True
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

unique_labels = set(labels)
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
for c, col in zip(unique_labels, colors):
    if c == -1:
        col = 'k'

    cl = (labels == c)

    xy = X[cl & cr]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col, markeredgecolor='k', markersize=14)

    xy = X[cl & ~cr]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col, markeredgecolor='k', markersize=6)

plt.title('Estimated number of clusters: %d' % n_clusters)

plt.subplot(222)
db = DBSCAN(eps=1, min_samples=2).fit(X)
labels = db.labels_
cr = np.zeros_like(labels, dtype = bool)
cr[db.core_sample_indices_] = True
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

unique_labels = set(labels)
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
for c, col in zip(unique_labels, colors):
    if c == -1:
        col = 'k'

    cl = (labels == c)

    xy = X[cl & cr]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col, markeredgecolor='k', markersize=14)

    xy = X[cl & ~cr]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col, markeredgecolor='k', markersize=6)

plt.title('Estimated number of clusters: %d' % n_clusters)

plt.subplot(223)
db = DBSCAN(eps=0.8, min_samples=10).fit(X)
labels = db.labels_
cr = np.zeros_like(labels, dtype = bool)
cr[db.core_sample_indices_] = True
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

unique_labels = set(labels)
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
for c, col in zip(unique_labels, colors):
    if c == -1:
        col = 'k'

    cl = (labels == c)

    xy = X[cl & cr]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col, markeredgecolor='k', markersize=14)

    xy = X[cl & ~cr]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col, markeredgecolor='k', markersize=6)

plt.title('Estimated number of clusters: %d' % n_clusters)

plt.subplot(224)
db = DBSCAN(eps=0.7, min_samples=5).fit(X)
labels = db.labels_
cr = np.zeros_like(labels, dtype = bool)
cr[db.core_sample_indices_] = True
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

unique_labels = set(labels)
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
for c, col in zip(unique_labels, colors):
    if c == -1:
        col = 'k'

    cl = (labels == c)

    xy = X[cl & cr]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col, markeredgecolor='k', markersize=14)

    xy = X[cl & ~cr]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col, markeredgecolor='k', markersize=6)

plt.title('Estimated number of clusters: %d' % n_clusters)
plt.show()
```

(c) Run the Agglomerative Clustering algorithm with several pairs of the affinity (the distance measures) and linkage (the linkage criteria). Examples include euclidean and
ward, and manhattan and complete.

```python
data = pd.read_csv('hwk09.csv')
h1 = data['x'].values
h2 = data['y'].values
X = np.array(list(zip(h1, h2)))
colors = np.array(['red', 'green', 'blue', 'orange', 'purple'])

fig = plt.figure(figsize=(10,10))

plt.subplot(221)
clustering = AgglomerativeClustering(linkage='ward', affinity='euclidean', n_clusters=5)
clustering.fit(X)
plt.scatter(X[:,0], X[:,1], s=10, c=[colors[i] for i in clustering.labels_])
plt.title("Linkage: Ward & Affinity: Euclidean")

plt.subplot(222)
clustering = AgglomerativeClustering(linkage='complete', affinity='manhattan', n_clusters=5)
clustering.fit(X)
plt.scatter(X[:,0], X[:,1], s=10, c=[colors[i] for i in clustering.labels_])
plt.title("Linkage: Complete & Affinity: Manhattan")

plt.subplot(223)
clustering = AgglomerativeClustering(linkage='average', affinity='l1', n_clusters=5)
clustering.fit(X)
plt.scatter(X[:,0], X[:,1], s=10, c=[colors[i] for i in clustering.labels_])
plt.title("Linkage: Average & Affinity: L1")

plt.subplot(224)
clustering = AgglomerativeClustering(linkage='single', affinity='cosine', n_clusters=5)
clustering.fit(X)
plt.scatter(X[:,0], X[:,1], s=10, c=[colors[i] for i in clustering.labels_])
plt.title("Linkage: Single & Affinity: Cosine")

plt.show()
```
