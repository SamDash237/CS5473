```python
import pandas as pd
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('hwk08.csv', delimiter=',')
```

```python
df.shape
```

```python
df1 = df.iloc[:, 0:2]
df2 = df.iloc[:, 2:4]
```

```python
kMeans1 = KMeans(n_clusters=2).fit(df1)
kMeans1_2 = KMeans(n_clusters=3).fit(df1)
kMeans1_3 = KMeans(n_clusters=4).fit(df1)
```

```python
plt.scatter(df1.iloc[:,0], df1.iloc[:,1], c=kMeans1.labels_, cmap='rainbow')
```

```python
plt.scatter(df1.iloc[:,0], df1.iloc[:,1], c=kMeans1_2.labels_, cmap='rainbow')
```

```python
plt.scatter(df1.iloc[:,0], df1.iloc[:,1], c=kMeans1_3.labels_, cmap='rainbow')
```

```python
kMeans2 = KMeans(n_clusters=2).fit(df2)
kMeans2_2 = KMeans(n_clusters=3).fit(df2)
kMeans2_3 = KMeans(n_clusters=4).fit(df2)
```

```python
plt.scatter(df2.iloc[:,0], df2.iloc[:,1], c=kMeans2.labels_, cmap='rainbow')
```

```python
plt.scatter(df2.iloc[:,0], df2.iloc[:,1], c=kMeans2_2.labels_, cmap='rainbow')
```

```python
plt.scatter(df2.iloc[:,0], df2.iloc[:,1], c=kMeans2_3.labels_, cmap='rainbow')
```
