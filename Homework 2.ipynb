(2.b.) Normalize the data points by making the norm of each data point (under columns C,
D, E, and F) equal to 1. That is, scale the values in columns C, D, E, and F, so that,
for each row (c, d, e, f) we have sqrt((c^2)+(d^2)+(e^2)+(f^2)) = 1. Then, print the row in the
DataFrame that has the shortest Euclidean distances from the normalized point p.

```python
for i in range(len(df) - 1):
    dfNorm = df[['C', 'D', 'E', 'F']].copy()
    normDenominator = math.sqrt(((df["C"].iloc[i])**2) + ((df["D"].iloc[i])**2) + ((df["E"].iloc[i])**2) + ((df["F"].iloc[i])**2))
    dfNorm["C"].iloc[i] = df["C"].iloc[i]/normDenominator
    dfNorm["D"].iloc[i] = df["D"].iloc[i]/normDenominator
    dfNorm["E"].iloc[i] = df["E"].iloc[i]/normDenominator
    dfNorm["F"].iloc[i] = df["F"].iloc[i]/normDenominator
    dfEuclideanNormalization = sp.spatial.distance.euclidean(dfNorm.iloc[1000], dfNorm.iloc[i])
    print(dfEuclideanNormalization)
```

(3.b.) Use the crosstab() function to construct the contingency table for columns A and B,
similar to the following sample, where the distinct values in attribute A are {a1, a2, a3}
and in attribute B are {b1, b2,} (this is just a sample and may not be the same as
the data in your data file). Write a sequence of Python code to perform the Pearsonâ€™s
chi-square (Ï‡2) test of independence with a confidence level of 0.001 to determine if
the two attributes are correlated. You should use stats.chi2() to get the Ï‡2 distribution.
Print sufficient information to report the result of the test.

```python
df = pd.read_csv(r'hwk01.csv', usecols=['A', 'B'])
```

```python
chi2TestResult = sp.stats.chi2_contingency(pd.crosstab(index=df['A'],columns=df['B']))
print (chi2TestResult)
if (chi2TestResult[1] >= .001):
    print ("Result is greater than confidence level of .001, therefore the two attributes are correlated.")
else:
    print ("Result is less than confidence level of .001, therefore the two attributes are not correlated.")
```
