```python
df2 = pd.read_csv('df2Final.csv')

train_X = df2[['department', 'age', 'salary']].to_dict('records')
train_y = df2['status']

def ageFormat(val):
    if 21 <= val <= 25:
        return '21..25'
    elif 26 <= val <= 30:
        return '26..30'
    elif 31 <= val <= 35:
        return '31..35'
    elif 36 <= val <= 40:
        return '36..40'
    elif 41 <= val <= 45:
        return '41..45'
    elif 46 <= val <= 50:
        return '46..50'

def salaryFormat(val):
    if 26 <= val <= 30:
        return '26k..30k'
    elif 31 <= val <= 35:
        return '31k..35k'
    elif 36 <= val <= 40:
        return '36k..40k'
    elif 41 <= val <= 45:
        return '41k..45k'
    elif 46 <= val <= 50:
        return '46k..50k'
    elif val <= 70:
        return '66k..70k'

df2Pipe = make_pipeline(DictVectorizer(), DecisionTreeClassifier())
df2Pipe.fit(train_X, train_y)

print('Enter department, age, and salary(with lowercase k at the end), separated by a space: ')
inputQ2 = input()
print()

inputQ2Split = inputQ2.split(' ')
ageDict = ageFormat(int(inputQ2Split[1]))
salaryDict = salaryFormat(int(inputQ2Split[2].rstrip('k')))
dataDict = {'department': inputQ2Split[0], 'age': ageDict, 'salary': salaryDict}

df2Pred = df2Pipe.predict(dataDict)

print('Decision Tree Classifier Prediction: {}:'.format(dataDict))
print(df2Pred[0])
```
